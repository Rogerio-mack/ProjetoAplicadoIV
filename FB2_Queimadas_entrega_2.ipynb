O modelo SARIMA (Auto Regressivos Integrados de Média Móveis com Sazonalidade) são bastante úteis para gerar previsões em Séries Temporais, quando não estão disponíveis variáveis preditoras. Pode ser descrito matematicamente como:

A diferença entre ARIMA e SARIMA está na inclusão do componente sazonal, ou seja, enquanto o ARIMA lida com dados em Séries Temporais não sazonais, o SARIMA modela e incorpora padrões sazonais nos dados. Outra diferença são os padrões adicionais que precisam ser estimados. No ARIMA, os parâmetros incluem a ordem autorregressiva(p), a ordem de diferenciação(d) e a ordem da média móvel(q). Já o SARIMA, os parâmetros incluem ordem autorregressiva sazonal(P), a ordem de diferenciação sazonal(D) e a ordem de média móvel sazonal(Q).
Como o  SARIMA  irá descrever e  analisar dados com previsões mais precisas e captura melhor os padrões em dados sazonais, utilizaremos  também, as funções de autocorrelação (ACF) e função de autocorrelação parcial (PACF) que são ferramentas importantes para a análise de Séries Temporais. A ACF que mede a correlação entre uma observação e seus valores defasados em vários períodos de tempo, ou seja, irá mostrar a relação dos focos de queimadas atual com as observações passadas. E a partir daí podemos tomar decisões mais assertivas. Quanto ao PACF, que nos ajuda a identificar a relação direta entre as observações de focos de queimadas de um modelo autorregressivo (AR), que indicará os termos defasados, mas que são significativos, ou seja, os menores focos mas que também impactam no meio ambiente.
